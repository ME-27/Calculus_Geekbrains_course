### –≤—ã–ø–æ–ª–Ω–∏–ª: –ï–≤–≥–µ–Ω–∏–π –ö–∏—Ä–∏–ª–ª–æ–≤

__1.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### $$\mbox{arctg}(\frac{y}{x})=\ln \sqrt{x^2+y^2}$$

### $$(\mbox{arctg}(\frac{y}{x}))'=(\ln \sqrt{x^2+y^2})'$$

### $$\frac{1}{1+(\frac{y}{x})^2}\cdot(\frac{y}{x})'=\frac{1}{\sqrt{x^2+y^2}}\cdot(\sqrt{x^2+y^2})'$$

### $$\frac{1}{1+(\frac{y}{x})^2}\cdot\frac{y'x-y}{x^2}=\frac{1}{\sqrt{x^2+y^2}}\cdot\frac{1}{2\sqrt{x^2+y^2}}\cdot(x^2+y^2)'$$

### $$\frac{y'x-y}{x^2+y^2}=\frac{1}{2(x^2+y^2)}\cdot(2x+2y\cdot y')$$

### $$\frac{y'x-y}{x^2+y^2}=\frac{2x+2yy'}{2(x^2+y^2)}$$

### $$y'x-y=x+yy'$$

### $$y'(x-y)=x+y$$

### $$y'=\frac{x+y}{x-y}$$

__2.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### \begin{cases} y=\frac{t^2}{t-1}, \\x=\frac{t}{t^2-1} \end{cases}

### \begin{cases} y'_{t}=(\frac{t^2}{t-1})'=\frac{(t^2)'(t-1)-t^2(t-1)'}{(t-1)^2}=\frac{2t^2-2t-t^2}{(t-1)^2}=\frac{t^2-2t}{(t-1)^2}, \\x'_{t}=(\frac{t}{t^2-1})'=\frac{(t)'(t^2-1)-t(t^2-1)'}{(t^2-1)^2}=\frac{t^2-1-2t^2}{(t^2-1)^2}=\frac{-t^2-1}{(t^2-1)^2} \end{cases}

###  $$y'_{x}=\frac{y'_{t}}{x'_{t}}=\frac{t^2-2t}{(t-1)^2} \cdot \frac{(t^2-1)^2}{-t^2-1}=\frac{t^4-3t^2-2t}{-t^2-1}$$

__3.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è:

### $$y=(x^2+2)^5\cdot(3x-x^3)^3$$

### $$y'=(x^2+2)^5\cdot(3x-x^3)^3(\ln{((x^2+2)^5\cdot(3x-x^3)^3)})'=$$

### $$=(x^2+2)^5\cdot(3x-x^3)^3((\ln{(x^2+2)^5)}'+(\ln{(3x-x^3)^3)'})=$$

### $$=(x^2+2)^5\cdot(3x-x^3)^3\cdot(\frac{5}{x^2+2}\cdot(x^2+2)' + \frac{3}{3x-x^3}\cdot(3x-x^3)')=$$

### $$=(x^2+2)^5\cdot(3x-x^3)^3\cdot(\frac{5}{x^2+2}\cdot2x + \frac{3}{3x-x^3}\cdot(3-3x^2))=$$

### $$=10x(x^2+2)^4(3x-x^3)^3 + (9-9x^2)(x^2+2)^5(3x-x^3)^2$$

–ø—Ä–æ–≤–µ—Ä–∏–º —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞

```python
from sympy import *
init_printing()
x = Symbol('x')
y = (x ** 2 + 2) ** 5 * (3 * x - x ** 3) ** 3
diff(y,x)
```

### $$y=x^x$$

### $$y'=x^x\cdot(\ln{x^x})'=x^x\cdot(x\ln{x})'=x^x\cdot((x)'\ln{x}+x(\ln{x})')=x^x\cdot(1\cdot\ln{x}+x\cdot\frac{1}{x})=x^x\cdot(\ln{x}+1)$$
