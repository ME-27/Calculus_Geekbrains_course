__1.__ Исследовать на условный экстремум функцию

### $$U=3-8x+6y,$$

если

### $$x^2+y^2=36$$

составим функцию Лагранжа:

### $$L(x,y,\lambda)=3-8x+6y +\lambda \cdot (x^2+y^2-36)$$

найдем частные производные, приравняем их к нулю и решим систему уравнений:

### $$\begin{cases}L'_{x}=-8+2\lambda x=0 \\L'_{y}=6+2\lambda y=0 \\L'_{\lambda}=x^2+y^2-36=0\end{cases}$$

### $$\Rightarrow \begin{cases} x=\frac{4}{\lambda} \\ y=\frac{3}{\lambda} \\ \lambda^2=\frac{25}{36}\end{cases}$$

#### $$\Rightarrow \left[ \begin{gathered} \left\{\begin{gathered}x=\frac{24}{5} \\ y=\frac{18}{5} \\ y=\frac{5}{6}\end{gathered}\right. \\ \left\{\begin{gathered}x=-\frac{24}{5} \\ y=-\frac{18}{5} \\ y=-\frac{5}{6} \end{gathered}\right.\end{gathered} \right.$$

получаем две точки $(\frac{24}{5},\frac{18}{5},\frac{5}{6})$ и $(-\frac{24}{5},-\frac{18}{5},-\frac{5}{6})$

найдём вторые производные:

### $$\begin{gathered}L''_{xx}=2\lambda \\ L''_{xy}=0 \\ L''_{x\lambda}=2x\end{gathered}$$

### $$\begin{gathered}L''_{yx}=0 \\ L''_{yy}=2\lambda \\ L''_{y\lambda}=2y\end{gathered}$$

### $$\begin{gathered}L''_{\lambda x}=2x \\ L''_{\lambda y}=2y \\ L''_{\lambda\lambda}=0\end{gathered}$$

далее найдем определитель в общем виде, а затем уже разберемся с каждой точкой:

### $$\Delta=\begin{vmatrix}2\lambda & 0 & 2x\\ 0 & 2\lambda & 2y\\2x & 2y & 0\\\end{vmatrix}=\begin{vmatrix}+ & - & +\\ - & + & -\\+ & - & +\\\end{vmatrix}=2\lambda \cdot\begin{vmatrix}2\lambda & 2y\\2y & 0\\\end{vmatrix}-0\cdot\begin{vmatrix}0 & 2y\\2x & 0\\\end{vmatrix}+2x\cdot\begin{vmatrix}0 & 2\lambda\\2x & 2y\\\end{vmatrix}\\=2\lambda \cdot (2\lambda \cdot0-2y\cdot2y)-0+2x\cdot(0\cdot2y - 2\lambda\cdot2x) = -8\lambda(x^2+y^2)$$

по условию $x^2+y^2=36$, тогда

### $$\Delta=-8\lambda \cdot 36=-288\lambda$$

т.е. знак определителя засисит только от знака $\lambda$

#### если $\lambda=\frac{5}{6}$, то $\Delta<0$, следовательно $(\frac{24}{5},\frac{18}{5},\frac{5}{6})$ - точка минимума

#### если $\lambda=-\frac{5}{6}$, то $\Delta>0$, следовательно $(-\frac{24}{5},-\frac{18}{5},-\frac{5}{6})$ - точка максимума

__3.__ Численно найти хотя бы один действительный корень системы нелинейных уравнений:
### $$\begin{cases}x^2-y^2+3xy^3-2x^2y^2+2x-3y-5=0 \\3y^3-2x^2+2x^3y-5x^2y^2+5=0\end{cases}$$

```python
from scipy.optimize import fsolve, broyden1
import math

def equations(p):
    x, y = p
    # Запись системы уравнения вида x+y=2, x-y=0
    return (x**2 - y**2 + 3*x*y**3 - 2*x**2*y**2 + 2*x - 3*y - 5, 3*y**3 - 2*x**2 + 2*x**3*y - 5*x**2*y**2 + 5)

# Численное решение системы уравнений
(x, y), info, ier, mesg =  fsolve(equations, (10, 10), full_output=True)
print(x, y, ier)
```

__4*.__ Численно найти все $5$ действительных корней.

```python
import numpy as np
from scipy.optimize import fsolve, broyden2
import math

def equations(p):
    x, y = p
    return (x**2 - y**2 + 3*x*y**3 - 2*x**2*y**2 + 2*x - 3*y - 5, 3*y**3 - 2*x**2 + 2*x**3*y - 5*x**2*y**2 + 5)

solv = set()
for x in np.arange(-100, 100, 1):
    for y in np.arange(-100, 100, 1):
        (x, y), info, ier, mesg =  fsolve(equations, (x, y), full_output=True)
        if (ier == 1): solv.add((int(x*10**8)/10**8, int(y*10**8)/10**8))
            
print(solv)
```
