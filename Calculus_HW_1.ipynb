### Задание 1

Представьте в виде __несократимой__ рациональной дроби:

### $$а)\,\,\, 0.(216);$$

### $$б)\,\,\, 1.0(01).$$

### <center>Решение:</center>

## $$а)$$
$$ 0.(216) = x$$
$$ x = 0.(216)$$
$$ 1000x = 216.(216)$$
$$ 1000x = 216 + 0.(216)$$
$$ 1000x = 216 + x$$
$$ 1000x - x = 216 + x - x$$
$$ 999x = 216$$
$$ x = \frac{216}{999} = \frac{\frac{216}{9}}{\frac{999}{9}} = \frac{24}{111}$$

## $$б)$$
$$ 1.0(01) = x$$
$$ x = 1.0(01)$$
$$ 1000x = 1001.0(01)$$
$$ 1000x = 1000 + 1.0(01)$$
$$ 1000x = 1000 + x$$
$$ 1000x - x = 1000 + x - x$$
$$ 999x = 1000$$
$$ x = \frac{1000}{999}$$

### Задание 6*

Прочитайте высказывания, установите их истинность и постройте противоположное высказывание:

### $$a)\,\,\, \forall x\in\mathbb{R}\,\,\,\exists X\in\mathbb{R}:\,\,\, X>x;$$
### $$б)\,\,\, \forall y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\exists \varepsilon>0:\,\,\, \sin y<\sin(y+\varepsilon);$$
### $$в)\,\,\, \forall y\in\Bigl[0; \pi\Bigr)\,\,\,\exists \varepsilon>0:\,\,\, \cos y>\cos(y+\varepsilon).$$

### <center>Решение:</center>

## <center>a)</center>

### $$\forall x\in\mathbb{R}\,\,\,\exists X\in\mathbb{R}:\,\,\, X>x$$

<center>выражение в словестном виде: для каждого числа $x$, принадлежащего множеству рациональных чисел, есть некое число $X$, принадлежащее множеству рациональных чисел, такое, что $X > x$.</center>

<center>выражение истинно, т.к. всегда найдётся рациональное число больше другого рационального числа. напишем противоположное выражение:</center>

### $$\exists x\in\mathbb{R}\,\,\,\forall X\in\mathbb{R}:\,\,\, X<x$$

<center>выражение в словестном виде: для некого числа $x$, принадлежащего множеству рациональных чисел при любом значении числа $X$, принадлежащих множеству рациональных чисел, выполняется условие $X < x$.</center>

<center>это противоположное выражение не является истинным.</center>

## <center>б)</center>

### $$\forall y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\exists \varepsilon>0:\,\,\, \sin y<\sin(y+\varepsilon)$$

<center>выражение в словестном виде: для каждого числа $y$, лежащего на отрезке от 0 до $\frac{\pi}{2}$, есть некое число $\varepsilon>0$, такое, что $\sin y$ < $\sin(y+\varepsilon)$.</center>

<center>выражение ложно, т.к. максимальным значением sin является $\sin(\frac{\pi}{2}) = 1$ и $\sin(y+\varepsilon)$ не может быть больше $\sin(\frac{\pi}{2})$ ни при каком значении $\varepsilon$.</center>

<center>напишем противоположное выражение:</center>

### $$\exists y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\forall \varepsilon>0:\,\,\, \sin y>\sin(y+\varepsilon)$$

<center>выражение в словестном виде: существует некий $y$, лежащий на отрезке от 0 до $\frac{\pi}{2}$, что для каждого числа число $\varepsilon>0$, ыполняется условие $\sin y>\sin(y+\varepsilon)$.</center>

<center>выражение истинно, т.к. максимальным значением sin является $\sin(\frac{\pi}{2}) = 1$, а соответственно sin от любого другого числа будет меньше 1.</center>

## <center>в)</center>

### $$\forall y\in\Bigl[0; \pi\Bigr)\,\,\,\exists \varepsilon>0:\,\,\, \cos y>\cos(y+\varepsilon)$$

<center>выражение в словестном виде: для каждого числа $y$, лежащего на отрезке от 0 до $\pi$ (исключительно), есть некое число $\varepsilon>0$, такое, что $\cos y>\cos(y+\varepsilon)$.</center>

<center>выражение истинно, т.к. минимальным значением cos является $\cos(\pi) = -1$, а т.к. $y$ в данном случае не может быть $\pi$, то всегда найдётся некое значение $\varepsilon$, которое увеличит значение $y + \varepsilon$ до $\pi$, а соответственно и $\cos(y+\varepsilon)$ будет меньше, чем $\cos(y)$.</center>

<center>напишем противоположное выражение:</center>

### $$\exists y\in\Bigl[0; \pi\Bigr)\,\,\,\forall \varepsilon>0:\,\,\, \cos y<\cos(y+\varepsilon)$$

<center>выражение в словестном виде: есть некое число $y$, лежащее на отрезке от 0 до $\pi$ (исключительно), что для каждого числа $\varepsilon>0$, выполняется условие $\cos y<\cos(y+\varepsilon)$.</center>

<center>выражение ложно, т.к. минимальным значением cos является $\cos(\pi) = -1$, а т.к. $y$ в данном случае не может быть $\pi$, то найдётся некое значение $\varepsilon$, которое увеличит значение $y + \varepsilon$ до $\pi$, т.е. до возможного абсолютного минимума.</center>
