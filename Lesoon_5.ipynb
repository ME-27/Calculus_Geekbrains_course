# Линейная функция
$$f(x)=kx+b, \ k,b \in \mathbb{R}$$

```python
import matplotlib.pyplot as plt
plt.axhline(y=0.5, color='r', linestyle='-')
plt.axhline(y=2, color='r', linestyle='-')
plt.axhline(y=4, color='r', linestyle='-')
plt.show()
```

```python
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-5,5,100)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, x, '-r', label='y=2x+1')
plt.plot(x, -2*x-5,'-g', label='y=-2x+5')
plt.plot(x, 2*x+1,'-b', label='y=2x+3')
plt.plot(x, 5*x-9,'-p', label='y=2x-3')

plt.legend(loc='upper left')
plt.show()
```

Коэффициент $k$ при x называется угловой коэффициент, он определяет, как быстро растет значение, b - свободный член, 
он определяет точку пересечения с осью ординат (значение при $x=0$). Точка пересечения с осью абсцисс будет b/k.

От значения k графически зависит как будет повернута кривая, какой угол составлять с осью абсцисс, возрастать или 
убывать на плюс бесконечности и т.д:

Более того, оказывается, что если рассмотреть тангенс угла между прямой и осью абсцисс, то он окажется равным k 
(то есть k влияет на угол прямой).  На рисунке ниже можно увидеть, что
$\tan{\alpha}=b/ (b/k)=k$ Для других случаев прямых и чуть других треугольников (как на рисунке выше) тоже будет получаться
это.

Это пригодится в производных так как в общем случае $\tan{\alpha}=(f(x_2)-f(x_1))/(x_2-x_1)$ , что по сути есть приращение
функции к приращению аргумента,

Таким образом, можно определить x^(-n) как 1/(x^n) или (1/x)^n

Дальше можно определить степень x^(1/n). Разберем случай x^(1/2). На прошлом вебинаре мы говорили про обратные функции и 
что функция g обратная к f это такая что f(g(x))=x (то есть в итоге x перевелся в x). В данном случае возникает вопрос, 
как найти обратную функцию к x^2. Например, 4 ^2=16 , а как определить обратную операцию. Ну и по определению
обратной функции к f , если f(x)=y, то f^(-1)(y)=x. То есть надо построить функцию, которая будет по каждому y
находить x^2=y (берем только положительный корень). Например, значением при x=4 у обратной функции будет такое число
a, что a^2=x. Это число 2. То есть f^(-1)(4)=2. Потому получаем функцию x^(1/2) (получается число 1/2 ,  то есть для  x^2
степень 2 , здесь как обратная операция 1/2).

```python
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(0,3,100)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])

plt.plot(x, x**(1/2),'-b', label='y=x^(1/2)')
plt.plot(x, x**(1/3),'-g', label='y=x^(1/4)')
plt.plot(x, x**(1/4),'-c', label='y=x^(1/6)')

plt.legend(loc='upper left')
plt.show()
```

По аналогии со степенной функцией можно ввести обратную функцию к a^x то есть это будет такое значение
которое равно той степени в которую надо возвести a чтобы получить x , то есть какое надо взять x чтобы
a^x=x. Например, какую степень у 3 надо взять чтобы получить 9 или 1/3. Это будут значения 2 и -1. Такая
функция называется логарифм, то есть log_3(9)=2, log_3(1/3)=-1. Получается функция log_a(x) такая что a^(log_a(x))=x
Так как логарифм  - обратная функция к показательной, то она работает на области отображения a^(x)  то есть
только на положительных числах. То есть логарифм на вход принимает значения только положительные x. К основанию
требования остаются те же что и у показательной, то есть быть больше 0 и не быть равной 1.

```python
import math
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(0.1,10,100)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
plt.plot(x, [math.log(y,2) for y in x],'-b', label='y=log_2(x)')
plt.plot(x, [math.log(y,(1/2)) for y in x],'-g', label='y=log_(1/2)(x)')
plt.legend(loc='upper left')
plt.show()
```

Если a>1 то при больших x log_a(x) это степень в которую надо возвести a>1 чтобы получить большое x. Ответом будет
большое положительное число , потому при x стремящемся к бесконечности log_a(x) стремится к бесконечности при a>1.
При x близких к 0 log_a(x) это степень в которую надо возвести a>1 чтобы получить близкое к нулю x. Например возвести 2 
в какую-то степень , чтобы получить 0.00001 Ответом будет какое-то отрицательное большое по модулю число, потому 
log_a(x) при x стремящемся к 0 стремится к минус бесконечности.
В случае a<1 будет наоборот (продумайте самостоятельно)

Известными основаниями для логарифма являются такие основания как 2, 10. Но оказалось более удобно использовать
