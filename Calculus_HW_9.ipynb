__1.__ Вычислить неопределенный интеграл

### $$\int \frac{2x+3}{(x-2)(x+5)}\,dx$$

т.к. $d((x-2)(x+5))=d(x^2+3x-10)=(x^2+3x-10)'dx=(2x+3)dx$, тогда:

$dx=\frac{d((x-2)(x+5))}{2x+3}$, тогда:

### $$\int \frac{2x+3}{(x-2)(x+5)}\,dx=\int \frac{2x+3}{(x-2)(x+5)}\frac{d((x-2)(x+5))}{2x+3}=\int \frac{d((x-2)(x+5))}{(x-2)(x+5)}$$

получается интеграл вида:

### $$\int \frac{\,dx}{x}=\ln |x|+C$$ 

тогда:

### $$\int \frac{d((x-2)(x+5))}{(x-2)(x+5)}=\ln |(x-2)(x+5)|+C$$

тоже самое, с помощью sympy:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = (2 * x + 3) / ((x - 2) * (x + 5))
f
```

```python
integrate(f, x)
```

всё правильно!

__2.__ Вычислить неопределенный интеграл

### $$\int e^{2x}\cos 3x\,dx$$

воспользуемся формулой интегрирования по частям:

### $$\int U\cdot dV=U\cdot V-\int V\cdot dU$$

где $U=e^{2x}$, а $dU=d(e^{2x})=(e^{2x})'dx=2e^{2x}dx$

а $dV=\cos(3x)dx$, соответственно: $V=\int cos(3x)dx=\frac{1}{3}\cdot \sin(3x)$

тогда:

### $$\int e^{2x}\cos 3x\,dx=e^{2x}\cdot \frac{1}{3}\sin 3x - \int \frac{1}{3}\cdot \sin 3x \cdot 2e^{2x}\,dx=e^{2x}\cdot \frac{1}{3}\sin 3x - \frac{2}{3}\int e^{2x}\cdot \sin 3x\,dx=$$

воспользуемся формулой интегрирования по частям ещё раз:

### $$=e^{2x}\cdot \frac{1}{3}\sin 3x - \frac{2}{3}(e^{2x}\cdot(-\frac{1}{3}\cos 3x) -\int 2e^{2x}\cdot (-\frac{1}{3})\cos 3x\,dx)=$$

### $$=\frac{3e^{2x}\sin 3x + 2e^{2x}\cos 3x}{9} -\frac{4}{9}\int e^{2x}\cos 3x\,dx$$

получаем уравнение:

### $$\int e^{2x}\cos 3x\,dx=\frac{3e^{2x}\sin 3x + 2e^{2x}\cos 3x}{9} -\frac{4}{9}\int e^{2x}\cos 3x\,dx$$

перенесём $-\frac{4}{9}\int e^{2x}\cos 3x\,dx$ в левую часть:

### $$\int e^{2x}\cos 3x\,dx+\frac{4}{9}\int e^{2x}\cos 3x\,dx=\frac{3e^{2x}\sin 3x + 2e^{2x}\cos 3x}{9}$$

### $$\frac{13}{9}\int e^{2x}\cos 3x\,dx=\frac{3e^{2x}\sin 3x + 2e^{2x}\cos 3x}{9}$$

### $$\int e^{2x}\cos 3x\,dx=\frac{3e^{2x}\sin 3x + 2e^{2x}\cos 3x}{13}$$

тоже самое, с помощью sympy:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = exp(2 * x) * cos(3 * x)
f
```

```python
integrate(f, x)
```

всё правильно!

__3.__ Вычислить определенный интеграл

### $$\int\limits_0^{\ln2} xe^{-x}\,dx$$

воспользуемся формулой интегрирования по частям:

### $$\int U\cdot dV=U\cdot V-\int V\cdot dU$$

где $U=x$, а $dU=dx$

а $dV=e^{-x}dx$, соответственно: $V=\int e^{-x}dx=-e^{-x}$

тогда:

### $$\int\limits_0^{\ln2} xe^{-x}\,dx=x\cdot (-e^{-x})\bigg|_0^{\ln2} -\int\limits_0^{\ln2} -e^{-x}\,dx=$$

### $$=(\ln2\cdot(-e^{-\ln2})-0\cdot(-e^0)) +\int\limits_0^{\ln2} e^{-x}\,dx=$$

### $$=-\frac{\ln2}{2} +e^{-x}\bigg|_0^{\ln2}=-\frac{\ln2}{2}+(-e^{-ln2}-(-e^0))=\frac{1}{2}-\frac{\ln2}{2}$$

тоже самое, с помощью sympy:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = x * exp(-x)
f
```

```python
integrate(f,(x, 0, ln(2)))
```

всё правильно!
