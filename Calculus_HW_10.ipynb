__1.__ Исследовать сходимость ряда. 


### $$\frac{1}{2\sqrt2}+\frac{1}{3\sqrt3}+\,...\,+\frac{1}{(n+1)\sqrt{n+1}}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)\sqrt{n+1}}$$

(*) Двумя различными признаками.

#### исследуем сходимость ряда с помощью второго признака сравнения:

### $$\frac{1}{2\sqrt2}+\frac{1}{3\sqrt3}+\,...\,+\frac{1}{(n+1)\sqrt{n+1}}+\,...\,\,\,\,\,\,\,\sim\,\,\,\,\,\,O\Bigr(\frac{1}{n^{1.5}}\Bigl)$$

т.к. степень равна $1.5>1$, то ряд сходится

#### исследуем сходимость ряда с помощью интегрального признака Коши:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = 1 / (x + 1) ** 1.5
f
```

```python
integrate(f,(x, 1, +oo))
```

интеграл посчитать получилось, следовательно ряд сходится

__2.__ Исследовать сходимость ряда

### $$\frac{1000}{1!}+\frac{1000^2}{2!}+\,...\,+\frac{1000^n}{n!}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{1000^n}{n!}$$

#### исследуем сходимость ряда с помощью интегрального признака Коши:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = 1000 ** x / factorial(x)
f
```

```python
integrate(f,(x, 1, +oo))
```

как видно sympy не может посчитать нужный нам интеграл в силу слабости алгоритма

#### исследуем сходимость ряда с помощью радикального признака Коши:

```python
from sympy import *
init_printing()
x=Symbol('x')
f = (1000 ** x / factorial(x)) ** (1/x)
f
```

```python
limit(f, x, +oo)
```

0<1, следовательно ряд сходится

#### исследуем сходимость ряда численно

т.к. 1000 в числителе является слишком большим числом для вычислений, сократим его до 10, что не скажется на факте сходимости или расходимости ряда - что будет верно для 10, то будет верно и для 1000!

```python
import math
sequence_sum = 0
for x in range(1, 50):
    y = (10 ** x) / (math.factorial(x))
    sequence_sum += y
    print(sequence_sum)
```

как видно, ряд не растёт бесконечно и сходится к определённому числу
