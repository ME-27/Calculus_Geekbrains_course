### $$^*\,\,\,\,\lim_{n\to \infty} \Bigl(\frac{1}{1\cdot2}+\frac{1}{2\cdot3}+\frac{1}{3\cdot4}+...+\frac{1}{(n-1)\cdot n}\Bigr)$$

найдём предел данной последовательности экспериментально, для чего напишем следующую функцию:

```python
def task_1_e_limit_search(n):
    x_1 = 1
    x_2 = 2
    sequence = []
    for i in range(n):
        sequence.append(1 / (x_1 * x_2))
        x_1 += 1
        x_2 += 1
    return sum(sequence)
```

посмотрим к какому числу стремится результат при увеличении числа n

```python
print(task_1_e_limit_search(10))
print(task_1_e_limit_search(100))
print(task_1_e_limit_search(1000))
print(task_1_e_limit_search(10000))
```

как видно, последовательность стремится к 1

### Задание 4

Пользуясь критерием Коши, докажите сходимость последовательности:

### $$a_n=\frac{\sin1}{2}+\frac{\sin2}{2^2}+\frac{\sin3}{2^3}+..\frac{\sin n}{2^n}\,\,\,\,\,\Rightarrow$$

### $$\Bigl\{a_n\Bigr\}_{n=1}^\infty=\Bigl\{\frac{\sin1}{2}, \frac{\sin1}{2}+\frac{\sin2}{2^2}, ..., a_n,...\Bigr\}$$


$^*$ Какой член последовательности можно взять в качестве предела с точностью $\varepsilon=10^{-7}$?

для решения напишем следующий код:

```python
import math


n = 1
epsilon = 10 ** -7
previous_element = 0
while True:
    current_element = previous_element + math.sin(n) / 2 ** n
    print(f'элемент последовательности №{n}: {current_element}')
    if n > 1:
        abs_of_difference = abs(previous_element - current_element)
        print(f'модуль разности между a({n - 1}) и a({n}) элементами: {abs_of_difference}')
        if abs_of_difference < epsilon:
            break
    
    previous_element = current_element
    n += 1
```

Как видно, цикл остановился на 22-м элементе, т.к. модуль разности был меньше значения эпсилон.
Но, из-за синуса в формуле, модуль разности между элементами сокращается не строго последовательно, а соответственно после 22 элемента могут быть модули разности элементов больше эпсилон. Внесём коррективы в наш код и посмотрим на динамику после 22-го элемента.

```python
import math


n = 1
epsilon = 10 ** -7
previous_element = 0
while n < 100:
    current_element = previous_element + math.sin(n) / 2 ** n
    print(f'элемент последовательности №{n}: {current_element}')
    if n > 1:
        abs_of_difference = abs(previous_element - current_element)
        print(f'модуль разности между a({n - 1}) и a({n}) элементами: {abs_of_difference}')
        if abs_of_difference < epsilon:
            print(f'модуль разности между a({n - 1}) и a({n}) элементами МЕНЬШЕ эпсилон!')

    previous_element = current_element
    n += 1
```

#### Ответ: как я понял, суть критерия сходимости последовательности Коши заключается в том, что по мере продвижения 'вправо' по элементам последовательности, расстояния между этими элементами на линии вещественных чисел - сужаются, а соответственно, у последовательности есть предел отличный от бесконечности. Последовательность, расмотренная в данном задании, пусть и не на каждом шаге, но сужается, и вырисовывается её предел - 0.5928376206979427. В качестве предела с гарантированной точностью $\varepsilon=10^{-7}$ можно взять 24-й член последовательности.

### Задание 5

Пользуясь критерием Коши, докажите расходимость последовательности:

### $$b_n=1+\frac{1}{2}+\frac{1}{3}+..\frac{1}{n}\,\,\,\,\,\Rightarrow$$

### $$\Bigl\{b_n\Bigr\}_{n=1}^\infty=\Bigl\{1, 1+\frac{1}{2},1+\frac{1}{2}+\frac{1}{3}, ..., b_n,...\Bigr\}$$

напишем следующую программу:

```python
n = 1
previous_element = 0
while n < 100:
    if n == 1:
        current_element = 1
    else:
        current_element = previous_element + 1 / n
        
    print(f'элемент №{n}: {current_element}')
    
    if n > 1:
        abs_of_difference = abs(previous_element - current_element)
        print(f'модуль разности между a({n - 1}) и a({n}) элементами: {abs_of_difference}')
        
    previous_element = current_element
    n += 1    
```

#### Как видно, несмотря на то, что расстрояние между элементами последовательности сужается, (а не растёт или остаётся неизменным, что является критерием расходимости Коши), последовательность продолжает рост в плюс бесконечность.
